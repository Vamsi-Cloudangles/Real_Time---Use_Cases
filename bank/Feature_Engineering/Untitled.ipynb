```python
import pandas as pd
```

```python
dataset = pd.read_csv('bank.csv')
```

```python
dataset.head()
```

```python
cate = []
nume = []
for col in dataset.columns:
    if dataset[col].dtype == object:
        cate.append(col)
    else:
        nume.append(col) 
```

```python
print(cate)
print(nume)
```

```python
n = int(input("Enter Howmany columns you need to groupby in 'Categorical Columns' "))
cate_col = list(map(str,input("Enter the names of groupby Columns from 'Categorical Columns'").strip().split()))[:n]
print(dataset.groupby(cate_col, as_index=False).mean())
```
