```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv('bank.csv')
```

```python
dataset.head()
```

```python
for i in dataset.columns:
    fig, ax = plt.subplots(1,1, figsize=(15, 6))
    sns.set_theme(style="whitegrid")
    sns.countplot(x = dataset[i][1:],data=dataset.iloc[1:], order=dataset[i][1:].value_counts().index, palette='Blues_r')
    #fig.text(0.1, 0.95, f'{dataset[i][0].split("(")[0]}', fontsize=16, fontweight='bold', fontfamily='serif')
    i = str(i)
    plt.xlabel(i, fontsize=20)
    plt.ylabel('count')
    plt.yticks(fontsize=13)
    plt.box(False)
```
